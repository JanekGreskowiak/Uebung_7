```python
import matplotlib.pyplot as plt
import numpy as np

tracer=np.loadtxt('tracer.txt',skiprows=1) # Daten
tracer = np.where(tracer == -999,np.nan,tracer)

x=2 # Abstand der Beobachtungsmessstelle zum Injektionsbrunnen in m
t=np.arange(0.1,40,0.1) # Zeit in Tagen
F = 1 # m2
ne = 0.25 
m = 200.0 # g/m3 oder mg/L

## Anpassungsparameter
v=0.0001 # Abstandsgeschwindigkeit m/d 
D=0.0001 # Dispersionskoeffizient m2/d

# Analytische Loesung 
Cmax=m/(ne*F*np.sqrt(4*np.pi*D*t))
C_tracer=Cmax*np.exp(-1/(4*D*t)*(x-v*t)**2)

plt.figure()
plt.plot(tracer[:,0],tracer[:,1],'ro',mec='k')
plt.plot(t,C_tracer,'b',lw=2,label='Tracer')
plt.xlim([0,40]);plt.xlabel('Zeit in Tagen')
plt.ylabel('Konzentration in mg/L');plt.legend(loc=1); plt.show()
```
